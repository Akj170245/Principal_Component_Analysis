```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
X, y = make_moons(n_samples = 1000,noise =0.01, random_state = 42)
plt.scatter(X[:,0],X[:,1],c = y,cmap = 'Spectral',alpha = 0.8)
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
x_pca = pca.fit_transform(X)
plt.title("PCA")
plt.scatter(x_pca[:,0],x_pca[:,1],c = y,cmap = 'Spectral')
plt.xlabel("component 1")
plt.ylabel("component 2")
plt.show()
```

```python
from sklearn.decomposition import KernelPCA
kpca = KernelPCA(kernel = 'rbf',gamma = 15)
k_pca = kpca.fit_transform(X)
plt.title("kernal PCA")
plt.scatter(k_pca[:,0],k_pca[:,1],c=y,cmap = 'Spectral')
plt.show()
```
